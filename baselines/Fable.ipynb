```python
import pandas as pd
from prophet import Prophet

trip_start_train = pd.read_pickle('trip_start_station_train_20200829-20220828.pkl')
trip_start_train = trip_start_train.rename(columns={"size": "y", "timestamp": "ds"})
trip_stop_train = pd.read_pickle('trip_stop_station_train_20200829-20220828.pkl')
trip_stop_train = trip_stop_train.rename(columns={"size": "y", "timestamp": "ds"})
trip_start_eval = pd.read_pickle('trip_start_station_eval_20220829-20220831.pkl')
trip_start_eval = trip_start_eval.rename(columns={"size": "y", "timestamp": "ds"})

trip_stop_eval = pd.read_pickle('trip_stop_station_eval_20220829-20220831.pkl')
trip_stop_eval = trip_stop_eval.rename(columns={"size": "y", "timestamp": "ds"})

```

```python
stations = trip_start_train['station'].unique()
big_pred_list = []
for i in stations: 
    series = trip_start_train[trip_start_train['station'] == i]
    m = Prophet()
    m.fit(series)
    forecast = m.predict(trip_start_eval[trip_start_eval['station'] == i])
    forecast['station']=i 
    big_pred_list.append(forecast)
    
big_preds_start = pd.concat(big_pred_list)
big_preds_start
```
